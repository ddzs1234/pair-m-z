## find z_do2>9.25 spectral 

```python
from shutil import copy2

from astropy.io import fits


def pick():
    with fits.open(
            '/Users/astro/Documents/notebooks/zs/pair-m-z/mz/z_do2_10_25.fits'
    ) as f:
        z_do2 = f[1].data.field('z_do2')
        plateifu = f[1].data.field('plateifu')
        for i in range(0, len(z_do2)):
            z_do2_1 = float(z_do2[i])
            pifu = plateifu[i]
            if z_do2_1 > 9.25:
                copy2(
                    '/Users/astro/Documents/notebooks/zs/pair-m-z/stacking/' +
                    pifu + '_1re_stack.jpg',
                    '/Users/astro/Documents/notebooks/zs/pair-m-z/z_do2_9_25/')


if __name__ == '__main__':
    pick()
```
